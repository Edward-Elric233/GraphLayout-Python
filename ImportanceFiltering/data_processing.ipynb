```python
node_type = origin_nodes.type.drop_duplicates()
node_type.index = list(range(node_type.size))
node_type
```

```python
tmp = pd.DataFrame(np.zeros((origin_nodes.shape[0], node_type.size + 2), dtype = int), columns = node_type.append(pd.Series(['All', 'Y']), ignore_index = True), index = origin_nodes.index,)
tmp
```

构造一个空的DataFrame，专门用来保存每个点的度

```python
nodes = origin_nodes.loc[:, ['name', 'type']].join(tmp)
nodes.index = nodes.name
nodes = nodes.loc[:, 'type':]
nodes
```

```python
def func_iter(edge):
    source = edge['source']
    source_type = nodes.loc[source, 'type']
    target = edge['target']
    target_type = nodes.loc[target, 'type']
    nodes.loc[source, target_type] += 1
    nodes.loc[source, 'All'] += 1
    nodes.loc[target, source_type] += 1
    nodes.loc[target, 'All'] += 1

#origin_links.head(5)
origin_links.apply(func_iter, axis=1)
nodes.head(10)
```

```python
nodes[nodes==0] = np.nan

def func_iter(node):
    ret = 0
    def func_node_type_iter(t):
        nonlocal ret
        if node[t] != node[t]:
            return
        else:
            tmp = node[t] / node['All']
            ret += tmp * tmp
    node_type.apply(func_node_type_iter)
    return ret
nodes.loc[:, 'Y'] = nodes.apply(func_iter, axis = 1)
nodes.head(10)
```

```python
grouped_type = nodes.groupby(by='type')
```

```python
type_data = grouped_type[node_type.append(pd.Series(['All']), ignore_index = True)].sum()
type_data['num'] = grouped_type['type'].count()
type_data
```

```python
for (t, df) in grouped_type:
    type_data.loc[t, 'Avg_K'] = type_data.loc[t, 'All'] / type_data.loc[t, 'num']
    tmp = df['All'] * df['All']
    type_data.loc[t, 'Avg_K_2'] = tmp.sum() / tmp.size
    type_data.loc[t, 'Avg_Y'] = df['Y'].sum() / df.shape[0]
    tmp = df['Y'] * df['Y']
    type_data.loc[t, 'Avg_Y_2'] = tmp.sum() / tmp.size

type_data[type_data == 0.0] = np.nan

for index, row in type_data.iterrows():
    type_data.loc[index, 'K_0'] = row[node_type].count()
    tmp = 0
    for i, v in row[node_type].items():
        if v != v:
            # v is nan
            continue
        tmp += (type_data.loc[i, 'num'] / type_data.loc[index, 'num']) ** 2
    type_data.loc[index, 'Y_r'] = tmp
        
    

type_data
```

```python
type_data['Delta_K'] = (type_data['Avg_K_2'] - type_data['Avg_K'] ** 2) ** 0.5 / type_data['K_0']
type_data['Delta_Y'] = (type_data['Avg_Y_2'] - type_data['Avg_Y'] ** 2) ** 0.5;
type_data['R'] = type_data['Avg_Y'] / type_data['Y_r']
type_data['Delta_R'] = type_data['Delta_Y'] / type_data['Y_r']

type_data
```

```python
output_nodes = np.log(type_data['R']*1e6)
output_nodes
```

```python
output_links = []
for idx, row in type_data[node_type].iterrows():
    for i, v in row.iteritems():
        if v != v:
            continue;
        output_links.append({'source':idx, 'target':i, 'value':v})

output_links = pd.DataFrame(output_links)
```
